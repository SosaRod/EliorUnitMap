```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Load your spreadsheet
file_path = 'Kelley Request.xlsx'
unit_df = pd.read_excel(file_path, sheet_name='Unit ')

# Initialize geocoder
geolocator = Nominatim(user_agent="unit_locator")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Prepare full addresses
unit_df['Full_Address'] = unit_df[['STREET ADDRESS', 'CITY', 'STATE', 'ZIP']].astype(str).agg(', '.join, axis=1)

# Geocode addresses
unit_df['Location'] = unit_df['Full_Address'].apply(geocode)

# Extract latitude and longitude
unit_df['Latitude'] = unit_df['Location'].apply(lambda loc: loc.latitude if loc else None)
unit_df['Longitude'] = unit_df['Location'].apply(lambda loc: loc.longitude if loc else None)

# Save results to CSV
unit_df.to_csv('geocoded_units.csv', index=False)
```
